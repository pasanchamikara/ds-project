## Frequency of requests for specific IPs on a Time series

As a means of predicting, we are clustering the logs (access and error) to 5 minute chunks, and analyze the number of requests received.

In order to consider, a specific request within a given specific time period to be malicious, considering the memory limits, we do expect it to with a frequency of more than 500 requests.


```python
# malicious = 1, not-malicious =0
ip_result_dataset = {}
for ip in ip_count_df_access.index:
    time_series = df_access[df_access['host'] == ip].resample('15min', on='time').host.count()
    time_series_df = pd.DataFrame({'time_period': time_series.index, 'request_count': time_series.values, 'is_malicious': (time_series.values > 500) })
    ip_result_dataset[ip] = time_series_df

ip_result_dataset
```

```python
import seaborn as sns

ip_malicious_probability = []
for idx, ip in enumerate(ip_result_dataset):
    plt.figure(idx)
    sns.lineplot(x='time_period', y='request_count', data=ip_result_dataset[ip], linewidth=3, color='purple').set(title='Time Series Plot for ' + ip)
    plt.xticks(rotation=90)
    ip_malicious_probability.append({"ip": ip, "malicious_probability": ip_result_dataset[ip][ip_result_dataset[ip]['is_malicious'] == True].shape[0] / ip_result_dataset[ip].shape[0]})

ip_malicious_probability
```

### Making use of Naive Bayes technique to predict maliciousness of a IP at a given time

```python
def get_ip_probability(ip):
    return df_access[df_access['host'] == str(ip)].shape[0] / df_access.shape[0]
```

```python
get_ip_probability("54.253.92.102")
```

```python
def get_ip_malicious_conditional_probability(is_malicious):
    total_malicious_sum = 0
    total_sum = 0 
    for record in ip_result_dataset:
        total_malicious_sum = total_malicious_sum + ip_result_dataset[record][ip_result_dataset[record]['is_malicious'] == True]["is_malicious"].count()
        total_sum = total_sum + ip_result_dataset[record]["is_malicious"].count()
    return total_malicious_sum / total_sum
```

```python
get_ip_malicious_conditional_probability(True)
```

```python
ip_result_dataset["54.253.92.102"][ip_result_dataset["54.253.92.102"]['is_malicious'] == True].count()
```

```python
ip_result_dataset["54.253.92.102"].count()
```

```python
malicious_probability = []
for ip in ip_count_df_access.index:
    malicious_probability.append({"ip": ip, "probability": (get_ip_probability(ip) * get_ip_malicious_conditional_probability(True))})
    prob_df = pd.DataFrame(malicious_probability)
    prob_df = prob_df[prob_df['probability'] > 0.001]
prob_df
# malicious_probability
```
